```python
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05+height,
                '%.1f GW' % float(height),
                ha='center', va='bottom')

plt.rcParams["figure.figsize"] = 9,6
n=8
ind=np.arange(n)
p1=plt.bar(ind,db['Hydropower Installed, IHA'],color='b',align='center')
p2=plt.bar(ind,db['Hydropower Development Potential'],bottom=db['Hydropower Installed, IHA'],color='r',align='center')
p3=plt.bar(ind,db['Hydropower Potential (GW)'],color='',align='center')
plt.ylabel('Hydropower Capacity (GW)')
plt.xticks(ind,db['Country'])
plt.title('Hydropower Capacity in Southeast Asia')
plt.legend((p1[0], p2[0]), ('Installed Capacity', 'Potential Capacity'),loc='upper left')
ax=plt.gca()
ax.set_ylim([0,80])
autolabel(p1)
autolabel(p3)


```
